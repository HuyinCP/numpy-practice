# Binomial Distribution

Phân phối nhị thức (Binomial Distribution) là một **phân phối xác suất rời rạc**, mô tả xác suất có đúng `k` lần thành công trong `n` lần thử độc lập, với mỗi lần thử có xác suất thành công là `p`.

---

## Đặc điểm

- Là phân phối rời rạc (Discrete).
- Kết quả của mỗi phép thử chỉ có 2 khả năng: **thành công** hoặc **thất bại** (Bernoulli trial).
- Các phép thử là **độc lập**.
- Xác suất thành công `p` là **không đổi** qua các lần thử.
- Biến ngẫu nhiên X: số lần thành công trong `n` lần thử.

---

## Công thức xác suất

Công thức xác suất để biến ngẫu nhiên `X` nhận giá trị `k` (có đúng `k` lần thành công trong `n` lần thử):

$$
    P(X = K) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Trong đó:
- `C(n, k) = n! / (k! * (n - k)!)`: tổ hợp chập `k` của `n`
- `n`: số lần thử (trials)
- `k`: số lần thành công
- `p`: xác suất thành công mỗi lần thử
- `1 - p`: xác suất thất bại

---

## Ví dụ minh họa

> Tung đồng xu 10 lần, xác suất ra mặt ngửa mỗi lần là 0.5. Xác suất để có đúng 6 lần mặt ngửa là bao nhiêu?

Áp dụng công thức:

```
n = 10, k = 6, p = 0.5

P(X = 6) = C(10, 6) * (0.5)^6 * (0.5)^4
         = 210 * (0.5)^10
         ≈ 0.205
```

---

## Các đặc trưng thống kê

| Đặc trưng              | Công thức                         |
|------------------------|-----------------------------------|
| Kỳ vọng (Mean)         | `E(X) = n * p`                    |
| Phương sai (Variance)  | `Var(X) = n * p * (1 - p)`        |
| Độ lệch chuẩn           | `σ = √(n * p * (1 - p))`          |
| Skewness               | `(1 - 2p) / √(n * p * (1 - p))`   |
| Kurtosis              | `(1 - 6p(1-p)) / (n * p * (1-p))` |

---
+ It has three parameters:

  + n - number of trials.

  + p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

  + size - The shape of the returned array.


````python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.displot(random.binomial(n=100, p=0.5, size=10))
````

````python
data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
````
